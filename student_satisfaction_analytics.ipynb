```python
import pandas as pd
import random

# Data referensi

prodi = [
    "Teknik Informatika",
    "Sistem Informasi",
    "Teknik Elektro",
    "Teknik Mesin",
    "Teknik Industri",
    "Teknik Sipil",
    "Arsitektur",
    "Kimia",
    "Manajemen",
    "Akuntansi",
    "Ilmu Komunikasi",
    "Hukum",
    "Psikologi",
    "Administrasi Publik",
    "Administrasi Bisnis",
    "Kesehatan Masyarakat",
    "Keperawatan",
    "Farmasi",
    "Kebidanan"
]

semester = [1, 2, 3, 4, 5, 6, 7, 8]

gender = [
    "Laki-laki",
    "Perempuan"
]

kepuasan = [
    "Sangat Puas",
    "Puas",
    "Cukup Puas",
    "Tidak Puas",
    "Sangat Tidak Puas"
]

perlu_ditingkatkan = [
    "Perpustakaan",
    "Kebersihan",
    "Laboratorium/Lab Komputer",
    "Ruang Kelas",
    "Sarana Ibadah"
]

saran = [
    "Perlu peningkatan fasilitas kampus.",
    "Kebersihan kampus sudah cukup baik.",
    "Laboratorium perlu diperbarui.",
    "Ruang kelas perlu lebih nyaman.",
    "Perpustakaan perlu menambah koleksi buku.",
    "Sarana ibadah perlu diperluas.",
    "Pelayanan kampus sudah cukup memuaskan."
]

# Membuat 100 data dummy

data = []

for i in range(100):
    data.append({
        "Program Studi": random.choice(prodi),
        "Semester Saat Ini": random.choice(semester),
        "Jenis Kelamin": random.choice(gender),
        "Kepuasan Perpustakaan": random.choice(kepuasan),
        "Kepuasan Kebersihan": random.choice(kepuasan),
        "Kepuasan Laboratorium": random.choice(kepuasan),
        "Kepuasan Ruang Kelas": random.choice(kepuasan),
        "Kepuasan Sarana Ibadah": random.choice(kepuasan),
        "Fasilitas yang Perlu Ditingkatkan": random.choice(perlu_ditingkatkan),
        "Saran": random.choice(saran)
    })

# Membuat DataFrame

df = pd.DataFrame(data)

# Menampilkan 10 data pertama

print(df.head(10))

# Simpan ke CSV

df.to_csv("student_satisfaction_analytics.csv", index=False)

print("\nFile berhasil dibuat: student_satisfaction_analytics.csv")
```

```python
mapping = {
    "Sangat Puas": 5,
    "Puas": 4,
    "Cukup Puas": 3,
    "Tidak Puas": 2,
    "Sangat Tidak Puas": 1
}
```

```python
kolom_kepuasan = [
    "Kepuasan Perpustakaan",
    "Kepuasan Kebersihan",
    "Kepuasan Laboratorium",
    "Kepuasan Ruang Kelas",
    "Kepuasan Sarana Ibadah"
]

for kolom in kolom_kepuasan:
    df[kolom] = df[kolom].map(mapping)
```
